```python
deaths = "../AQI CSV Files_raw_data/covid19_deaths_global.csv"
deaths = pd.read_csv(deaths, delimiter=",")
```

```python
reduced_df = deaths.drop(columns=['Lat', "Long","Province/State"])
reduced_df 
```

```python
# drop_df = reduced_df.drop(columns=["Province/State"])
# drop_df
```

```python
jan_sum = reduced_df.iloc[:,2:12].sum(axis=1)
feb_sum = reduced_df.iloc[:,12:41].sum(axis=1)
mar_sum = reduced_df.iloc[:,41:72].sum(axis=1)
apr_sum = reduced_df.iloc[:,72:102].sum(axis=1)
may_sum = reduced_df.iloc[:,102:133].sum(axis=1)
jun_sum = reduced_df.iloc[:,133:163].sum(axis=1)
jul_sum = reduced_df.iloc[:,163:186].sum(axis=1)
reduced_df["Jan_sum"] = jan_sum
reduced_df["Feb_sum"] = feb_sum
reduced_df["Mar_sum"] = mar_sum
reduced_df["Apr_sum"] = apr_sum
reduced_df["May_sum"] = may_sum
reduced_df["Jun_sum"] = jun_sum
reduced_df["Jul_sum"] = jul_sum
reduced_df
```

```python
# reduced_df.iloc[:,2:186]])
totals_dff = reduced_df.drop(reduced_df.iloc[:,2:185], axis=1)
totals_df = totals_dff.rename(columns={"Country/Region":"Country"})
totals_new = totals_df[["Country","Jan_sum","Feb_sum","Mar_sum","Apr_sum","May_sum","Jun_sum","Jul_sum"]]
totals_new
```

```python
country = totals_new.groupby(["Country"]).sum()
col_names = country.rename(columns={'Jan_sum':'Jan', 'Feb_sum':'Feb', 'Mar_sum':'Mar', 'Apr_sum':'Apr', 'May_sum':'May',
       'Jun_sum':'Jun', 'Jul_sum':'Jul'})
covid_deaths_df = col_names.rename({"Korea, South":"South Korea"}, axis='index')
covid_deaths_df
```
