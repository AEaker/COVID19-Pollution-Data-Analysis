### Import Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
### Read in Los Angeles CSV
```

```python
la_df = pd.read_csv("Resources/la_months.csv")
la_df.head()
```

### Separating out the years to combine in Summary_Data_and_plotting.ipynb

* Leaving out 2014 for heatmap purpose

#### 2015

```python
la_15 = la_df.loc[(la_df["Year"]==2015)]
la_15
```

#### 2016

```python
la_16 = la_df.loc[(la_df["Year"]==2016)]
la_16
```

### Merge 2015 and 2016

```python
la_1516 = pd.merge(la_15,la_16,how='outer',on=["Month"],suffixes=("_15","_16"))
la_1516
```

#### 2017

```python
la_17 = la_df.loc[(la_df["Year"]==2017)]
la_17
```

#### 2018

```python
la_18 = la_df.loc[(la_df["Year"]==2018)]
la_18
```

### Merge 2017 and 2018

```python
la_1718 = pd.merge(la_17,la_18,how='outer',on=["Month"],suffixes=("_17","_18"))
la_1718
```

### Merge 2015/16 and 2017/18

```python
la_1518 = pd.merge(la_1516,la_1718,how='outer',on=["Month"])
la_1518
```

#### 2019

```python
la_19 = la_df.loc[(la_df["Year"]==2019)]
la_19
```

#### 2020

```python
la_20 = la_df.loc[(la_df["Year"]==2020)]
la_20
```

### Merge 2019 and 2020

```python
la_1920 = pd.merge(la_19,la_20,how='outer',on=["Month"],suffixes=("_19","_20"))
la_1920
```

### Merge 2015-18 and 2019/20

```python
la_all = pd.merge(la_1518,la_1920,how='outer',on=["Month"])
la_all
```

### Dropping the year columns no longer needed

```python
la_all2 = la_all.drop(["Year_15","Year_16","Year_17","Year_18","Year_19","Year_20"],axis=1)
la_all2
```

### Renaming the columns

```python
los_angeles = la_all2.rename(columns={
    
                                    "AQI_15":"Avg AQI_15",
                                    "AQI_16":"Avg AQI_16",
                                    "AQI_17":"Avg AQI_17",
                                    "AQI_18":"Avg AQI_18",
                                    "AQI_19":"Avg AQI_19",
                                    "AQI_20":"Avg AQI_20"
        
                                    })
los_angeles
```

```python
los_angeles.to_csv("output_data/los_angeles.csv",index=False)
```
